# 1. Loading and Filtering Dataset

I chose Barcelona as my city

```python
# Assigning the csv's to a data frame
df1 = pd.read_csv("https://osf.io/utwjs/download")
df2 = pd.read_csv("https://osf.io/p6tyr/download")


# Combining the both data frames
# used outer join to ensure every hotel id is merged into a new file.

df = df1.merge(df2, how = 'outer', on = "hotel_id")
```

```python
df['city'].value_counts().head(10)
```

 Barcelona has 5131 observations without any filters


```python
# filtering for Barcelona hotels
    
df = df[(df['city'] == 'Barcelona')&(df['accommodation_type'] == 'Hotel')]
```

```python
# Creating a binary highly_rated columns 

df["highly_rated"] = 0
df.loc[df["rating"] >= 4, "highly_rated"] = 1 
```

```python
# Dropping null values in any columns
df.dropna(inplace=True)
```
